```python
# -------------------------------------------------------------------
# CELL 1: Load API Key
# -------------------------------------------------------------------
# os        → built-in Python module to interact with the operating system
# dotenv    → reads the .env file and loads variables into the environment
# load_dotenv() scans for a .env file and loads its contents as 
# environment variables so we can access them securely via os.getenv()
# -------------------------------------------------------------------

import os
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Retrieve the API key (never print the full key!)
api_key = os.getenv("OPENAI_API_KEY")
print("API key loaded:", api_key[:8] + "...")
```

```python
# -------------------------------------------------------------------
# CELL 2: Test OpenAI API Connection
# -------------------------------------------------------------------
# OpenAI  → official Python client for the OpenAI API
# We initialize the client (it automatically picks up the API key 
# from the environment variables loaded in Cell 1)
# Then we send a simple test message to confirm the connection works
# -------------------------------------------------------------------

from openai import OpenAI

# Initialize the OpenAI client
client = OpenAI()

# Send a simple test message to GPT-4o-mini
response = client.chat.completions.create(
    model="gpt-4o-mini",        # the model we'll use throughout this project
    messages=[
        {"role": "user", "content": "Say hello in one sentence!"}
    ]
)

# Print the response
print(response.choices[0].message.content)
```

```python
# -------------------------------------------------------------------
# CELL 3: Test Chinook Database Connection
# -------------------------------------------------------------------
# sqlite3 → Python's built-in library for working with SQLite databases
# We connect to the Chinook database file and run a simple test query
# to confirm the database is accessible and working correctly
# -------------------------------------------------------------------

import sqlite3

# Connect to the Chinook SQLite database
conn = sqlite3.connect("Chinook_Sqlite.sqlite")
cursor = conn.cursor()

# Run a simple test query
cursor.execute("SELECT * FROM Artist LIMIT 10")
results = cursor.fetchall()

print("Database connection successful! Sample artists:")
for row in results:
    print(row)
```
